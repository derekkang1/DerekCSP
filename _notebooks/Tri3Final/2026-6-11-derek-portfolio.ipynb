### CS Project 1: OCS GitHub Analytics & Grade Predictor
*AP Computer Science A · Backend Engineer · Flask · GitHub GraphQL API · JWT Auth*

As part of a class project, I worked with a team to expand the Open Coding Society (a class-wide site) dashboard into a better analytics platform. As backend engineer, I tested and hardened all existing analytics endpoints in Postman, standardized admin routes to `GET` for REST consistency, and added line-change metrics to the commit stats model.

**My contributions:**
- Tested all analytics API endpoints end-to-end via Postman and frontend integration
- Converted admin endpoints from `POST` to `GET` for REST compliance
- Extended the commit stats function to also return per-commit line additions, deletions, and net totals — giving a more complete picture of contribution volume
- Built an admin frontend panel that looks up any user's commits by UID and renders them as linked commit cards
- Integrated with the OCS dashboard merge, coordinating with peer grade-analytics groups

The result was a significantly more useful dashboard: instructors can drill into any student's commit history at a glance, and the grade predictor now accounts for code volume in addition to commit count.

### CS Project 2: InterTravel – Budget Brilliance Module
*AP Computer Science A · Full-Stack · Flask · SQLAlchemy · JavaScript*

This was another project I did in class with a small team of 6. We worked together to create a travel planning site, and I did a specific aspect of it.

Built the budgeting subsystem for a group travel-planner app serving 5 international cities. Users enter a trip budget, lock in known expenses, then allocate remaining funds across categories. The system persists entries per user to a backend database and updates the remaining-budget display in real time.

I built the full budgeting database model, JWT-secured CRUD API, and the frontend table with a live remaining-budget counter. The API persists each user's expense entries to a backend database and the frontend fetches and re-renders the table after every change. I also integrated a currency converter API so users could budget in their local currency, and customized the Flask admin backend UI to reflect the InterTravel brand.

### Developing an ESP32-CAM Edge AI System for Bicycle Rearward Vehicle Detection
*Engineering: Energy, Materials, and Transport*

I built a simple and affordable device that would alert bicyclists of cars approaching them form behind in order to increase biker awareness on public roads. The existing solutions were all expensive radar based so I decided to use a relatively cheap ESP32 microcontroller and camera to create a vision model that would alert just as well. 

I designed and trained an AI model with images taken on the camerea and optimized the model to work on the contrained hardware.
