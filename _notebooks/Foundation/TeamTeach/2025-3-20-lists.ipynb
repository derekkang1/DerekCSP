### Introduction to Python Lists video:
- Lists are a built-in data structure in Python designed for storing and accessing objects in a specific sequence.
- Order matters in lists, making them ideal for working with ordered data.
- Lists can be created using the list() constructor or square brackets ([]), with the latter being simpler and more common.
- Lists can be pre-populated with values or updated later using methods like append().
- Items in a list are accessed by their index, starting from 0 (zero-based indexing).
- Negative indices allow access to elements from the end of the list, with -1 representing the last item.
- Python lists can contain multiple data types (e.g., integers, strings, booleans, floats) within the same list, unlike some other programming languages that require uniform types

## Introduction to List Comprehensions
- List comprehensions provide a concise and readable way to create lists compared to traditional for loops. 
- They can simplify code significantly while maintaining clarity.
- List comprehensions are often more readable and intuitive than using map and filter functions, making them a preferred choice for transforming or filtering data in Python.
    - example: creating a list of squares (n * n for n in nums) and filtering even numbers (n for n in nums if n % 2 == 0).
- The video also covers dictionary comprehensions (e.g., {name: hero for name, hero in zip(names, heroes)}) and set comprehensions (e.g., {n for n in nums})
- generator expressions are introduced as an efficient alternative to list comprehensions when working with large datasets, as they yield items lazily instead of creating a full list in memory.

## Homework Hack #2

```python
original_list = list(range(1, 31))
filtered_list = [num for num in original_list if num % 3 == 0 and num % 5 != 0]

print("Original List (1 to 30):")
print(original_list)

print("Filtered List:")
print(filtered_list)
```

```python
import pandas as pd

def filter_spotify_data(csv):
    spotify_data = pd.read_csv(csv)

    count = 0

    for a in spotify_data["Total Streams (Millions)"]:
        if a >= 10:
            print(str(spotify_data["Album"][count]), "has", str(a), "million streams")
            
        count += 1
        
filter_spotify_data("Spotify_2024_Global_Streaming_Data.csv")
```
